```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



ordinal_features = [
    'xyz_concern', 'xyz_knowledge', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
    'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective', 'opinion_seas_risk', 
    'opinion_seas_sick_from_vacc'
]

#  LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to each ordinal feature
for feature in ordinal_features:
    df[feature] = label_encoder.fit_transform(df[feature])
    df2[feature] = label_encoder.fit_transform(df2[feature])
    
# Apply One-Hot Encoding to nominal variables
df = pd.get_dummies(df, columns=nominal_features, drop_first=True)
df2 = pd.get_dummies(df2, columns=nominal_features, drop_first=True)


```
